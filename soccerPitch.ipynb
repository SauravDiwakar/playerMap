```python
"first install mplsoccer library(if you haven't) using - !pip install mplsoccer"
from mplsoccer import Pitch
import matplotlib.pyplot as plt

pitch = Pitch(pitch_color='grass', line_color='white',stripe=True)
fig, ax = pitch.draw()
plt.show()
```
